```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


pc = pd.read_csv('https://goo.gl/g8y9TC')
pc['Masa_Invariante'] = ((pc.E1+pc.E2)**2 - ((pc.px1+pc.px2)**2 + (pc.py1+pc.py2)**2 + (pc.pz1+pc.pz2)**2))**.5
pc.head()
len(pc)

fig = plt.figure(figsize=(16, 12))
plt.hist(pc.Masa_Invariante, 500, range = (1.5,120), facecolor='g')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()


fig = plt.figure(figsize=(16, 12))
plt.hist(pc.Masa_Invariante, 500, range = (1.5,5.25), facecolor='g')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()

jpsy = pc[(pc.Masa_Invariante >= 2.9) & (pc.Masa_Invariante <= 3.5)]
xc0 = pc[(pc.Masa_Invariante >= 3.5) & (pc.Masa_Invariante <= 4)]
fig = plt.figure(figsize=(15, 10))

plt.hist(jpsy.Masa_Invariante, 500, range = (2.8,4.1), facecolor='r')
plt.hist(xc0.Masa_Invariante, 500, range = (2.8,4.1), facecolor='g')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()

jpsy = pc[(pc.Masa_Invariante >= 2.9) & (pc.Masa_Invariante <= 3.5)]
fig = plt.figure(figsize=(16,9))

plt.hist(jpsy.Masa_Invariante, 600, range = (2.8,3.6), facecolor='r')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()

xc0 = pc[(pc.Masa_Invariante >= 3.5) & (pc.Masa_Invariante <= 4)]
fig = plt.figure(figsize=(16,9))

plt.hist(xc0.Masa_Invariante, 400, range = (3.4,4.1), facecolor='g')
plt.xlabel('Invariant mass (GeV/$c^2$)', fontsize = 15)
plt.ylabel('Number of events \n', fontsize = 15)
plt.title('Measured distribution of events with two muons \n', fontsize = 15)

plt.show()
```
